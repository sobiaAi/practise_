```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
A = np.array([[0.5 , 0.5, 0.25],[0.25 , 0, 0.25],[0.25 , 0.5, 0.5]])
print(A)
```

```python
Xtod=np.array([[1],[0],[0]])
print(Xtod)
```

```python
print(A@Xtod)
```

```python
weather= np.zeros((3,50))
weather[:,0]= Xtod[:,0]
for k in range(1,50):
    xtom= A@Xtod
    weather[:,k]= xtom[:,0]
    print(k)
    print(xtom)
    Xtod= xtom
```

```python
plt.figure(figsize=(8,5))
plt.plot(weather.T, linewidth=3)
plt.grid(True)
plt.ylim(0.18,0.5)
```
